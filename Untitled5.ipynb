```python
import socket
from datetime import datetime

# MISSION: Detect and Identify unauthorized access (The Digital Tripwire)
def digital_tripwire(host='0.0.0.0', port=8080):
    # Set up the "Fake" Service
    trap = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    trap, socket.PACKET_OTHERHOST
    trap.listen(5)

    print(f"[*] Sherlock's Trap active on port {port}. Watching for 'Suspects'...")

    while True:
        intruder, addr = trap.accept()
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

        # Log the Suspect's Identity
        print(f"\n[!] ALERT: Intruder detected from {addr[0]} at {timestamp}")

        # Forensic Capture of the "Payload" (the data they sent)
        payload = intruder.recv(1024).decode('utf-8', errors='ignore')

        # Record everything in a "Case File"
        with open("forensic_log.txt", "a") as case_file:
            case_file.write(f"--- SUSPECT FOUND ---\nIP: {addr[0]}\nTime: {timestamp}\nData: {payload}\n\n")

        # Send a standard error to avoid suspicion
        intruder.send(b"HTTP/1.1 200 OK\r\n\r\nHello Friend. You've been logged.")
        intruder.close()

if __name__ == "__main__":
    digital_tripwire()
```
